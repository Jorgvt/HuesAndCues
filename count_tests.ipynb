```python
from glob import glob
from collections import defaultdict
import pandas as pd
```

```python
p_lims = [0.05, 0.003]
```

```python
counts = defaultdict(dict)
for path in glob("./Results/*/p_values.csv"):
    model_name = path.split("/")[-2]
    df = pd.read_csv(path, index_col=0)
    for p_lim in p_lims:
        df_ = df[df["p-value"] < p_lim]
        counts[model_name][p_lim] = len(df_)/len(df)
        # break
    # break
```

```python
counts = pd.DataFrame.from_dict(counts, orient="index")
counts = counts.reset_index(names="model")
counts.head()
```

```python
def split_name(row):
    name = row.model
    dst = name.split("-")[-1]
    name = "-".join(name.split("-")[:-1])
    return {"model": name, "dataset": dst}
```

```python
counts = pd.concat([counts.apply(split_name, result_type="expand", axis=1), counts.drop("model", axis=1)], axis=1)
counts.head()
```

```python
counts.to_csv("pass_test_results.csv", index=0)
```
