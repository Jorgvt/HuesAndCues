```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("pass_test_results.csv")
df.head()
```

```python
df.drop("dataset", axis=1)\
  .groupby(by="model")\
  .mean()\
  .reset_index()\
  .plot(x="model", y=["0.003", "0.05"], kind="bar", figsize=(20,5))
plt.ylabel("No pasan el test (ratio)")
plt.axhline(0.5, linestyle="--", color="green")
plt.axhline(0.25, linestyle="--", color="red")
plt.show()
```

```python
df.drop("model", axis=1)\
  .groupby(by="dataset")\
  .mean()\
  .reset_index()\
  .plot(x="dataset", y=["0.003", "0.05"], kind="bar", figsize=(20,5))
plt.ylabel("No pasan el test (ratio)")
plt.axhline(0.5, linestyle="--", color="green")
plt.axhline(0.25, linestyle="--", color="red")
plt.show()
```

```python
df[df.dataset.str.match("laion")]\
  .drop("model", axis=1)\
  .groupby(by="dataset")\
  .mean()\
  .reset_index()\
  .plot(x="dataset", y=["0.003", "0.05"], kind="bar", figsize=(20,5))
plt.ylabel("No pasan el test (ratio)")
plt.axhline(0.5, linestyle="--", color="green")
plt.axhline(0.25, linestyle="--", color="red")
plt.show()
```
